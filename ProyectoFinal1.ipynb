```python
import pandas as pd
archivo_csv = "df_sucio.csv"
df = pd.read_csv(archivo_csv)
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
columnas_a_limpiar = [
    "Attendance",
    "Sleep_Hours",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Physical_Activity",
    "Exam_Score","Hours_Studied",]
```

```python
for columna in columnas_a_limpiar:
    df[columna] = df[columna].replace(" ", 0)
    df[columna] = df[columna].fillna(0)
```

```python
colText = [
    "Parental_Involvement",
    "Access_to_Resources",
    "Extracurricular_Activities",
    "Motivation_Level",
    "Internet_Access",
    "Teacher_Quality",
    "School_Type",
    "Peer_Influence",
    "Learning_Disabilities",
    "Parental_Education_Level",
    "Gender","Distance_from_Home","Family_Income"
]
```

```python
for columna in colText:
    df[columna] = df[columna].fillna("Desconocido")
    df[columna] = df[columna].str.strip()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.replace("Hours_Studied", "Horas_Estudio")
df.columns = df.columns.str.replace("Attendance", "Asistencia")
df.columns = df.columns.str.replace("Parental_Involvement", "Involucramiento_Padres")
df.columns = df.columns.str.replace("Access_to_Resources", "Acceso_a_Recursos")
df.columns = df.columns.str.replace("Extracurricular_Activities", "Actividades_Extracurriculares")
df.columns = df.columns.str.replace("Sleep_Hours", "Horas_Sueño")
df.columns = df.columns.str.replace("Previous_Scores", "Puntajes_Anteriores")
df.columns = df.columns.str.replace("Motivation_Level", "Nivel_Motivacion")
df.columns = df.columns.str.replace("Internet_Access", "Acceso_a_Internet")
df.columns = df.columns.str.replace("Tutoring_Sessions", "Sesiones_Tutorias")
df.columns = df.columns.str.replace("Family_Income", "Ingreso_Familiar")
df.columns = df.columns.str.replace("Teacher_Quality", "Calidad_Maestro")
df.columns = df.columns.str.replace("School_Type", "Tipo_Escuela")
df.columns = df.columns.str.replace("Peer_Influence", "Influencias_Compañeros")
df.columns = df.columns.str.replace("Physical_Activity", "Actividad_Fisica")
df.columns = df.columns.str.replace("Learning_Disabilities", "Discapacidad_de_Aprendizaje")
df.columns = df.columns.str.replace("Parental_Education_Level", "Nivel_Educacion_Padres")
df.columns = df.columns.str.replace("Distance_from_Home", "Distancia_Hogar")
df.columns = df.columns.str.replace("Gender", "Genero")
df.columns = df.columns.str.replace("Exam_Score", "Puntaje_Examen")
```

```python
df['Genero'] = df['Genero'].replace({'Male': 'Masculino', 'Female': 'Femenino'})
df['Nivel_Motivacion'] = df['Nivel_Motivacion'].replace({'High': 'Alta', 'Low': 'Baja', 'Medium': 'Media'})
df['Calidad_Maestro'] = df['Calidad_Maestro'].replace({'High': 'Alta', 'Low': 'Baja', 'Medium': 'Media'})
df['Ingreso_Familiar'] = df['Ingreso_Familiar'].replace({'High': 'Alta', 'Low': 'Baja', 'Medium': 'Media'})
df['Acceso_a_Recursos'] = df['Acceso_a_Recursos'].replace({'High': 'Alta', 'Low': 'Baja', 'Medium': 'Media'})
df['Involucramiento_Padres'] = df['Involucramiento_Padres'].replace({'High': 'Alta', 'Low': 'Baja', 'Medium': 'Media'})
df['Distancia_Hogar'] = df['Distancia_Hogar'].replace({'Far': 'Lejos', 'Near': 'Cerca', 'Moderate': 'Moderado'})
df['Influencias_Compañeros'] = df['Influencias_Compañeros'].replace({'Positive': 'Positivo', 'Negative': 'Negativa'})
df['Tipo_Escuela'] = df['Tipo_Escuela'].replace({'Public': 'Publica', 'Private': 'Privada'})
df['Acceso_a_Internet'] = df['Acceso_a_Internet'].replace({'Yes': 'Si'})
df['Discapacidad_de_Aprendizaje'] = df['Discapacidad_de_Aprendizaje'].replace({'Yes': 'Si'})
df['Actividades_Extracurriculares'] = df['Actividades_Extracurriculares'].replace({'Yes': 'Si'})
df['Nivel_Educacion_Padres'] = df['Nivel_Educacion_Padres'].replace({'College': 'Universidad', 'High School': 'Preparatoria', 'Postgraduate': 'Posgrado'})


```
